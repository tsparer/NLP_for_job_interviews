```python
# Note on conventions used below:  "list of potentials" is a list of leads and 
# follow up conversations generated by the bot:
# Each list contains a record (JSON/python dictionary style) of a conversation and a some 
# meta data (whether or not the admins thought the person was qualified based on the 
# conversation, the specific job posting tag, etc.).  
#  So: "list of potentials" is the entire list of records, "record" is an individual
# conversation and related data.
# (NOTE: the convos are from multiple sources/times, and in JSON- so the initial formatting
# and tracking systems do not always align.  Thus the necessity of lots of cleaning functions)



########################### Functions for + cleaning getting applicant replies #####
def get_applicant_replies(curr_discussion, applicant_auth_ident) 
""""takes a discussion with a bot, and the name of the the human author, and extracts
   only the human's replies.  Returns a list of strings of those replies."""
        human_replies = []                   
        for speaker_line in curr_discussion:
            if speaker_line["author_ident"] == applicant_auth_ident:
                applicant_response = speaker_line["content"]
                human_replies.append(applicant_response)
        return(human_replies)

def clean_human_replies()
    """removes extranaous formatting on human replies to bots, reduces list of strings
    to a single string, contained in a list. (The "document" form to be used for LDA etc) """


def get_clean_replies(curr_discussion, applicant_auth_ident):
    replies         = get_applicant_replies(curr_discussion, applicant_auth_ident)
    cleaned_replies = clean_human_replies(replies)
    return(cleaned_replies)

#############################################

def get_admin_score(record):
    """"checks an individual record to see if there is a note from the admin
    if there is, returns the score.  If not, returns "MISSING"""
        if "admin_insights" in record:
            gen_insight = record['admin_insights']
            if "admin_published_score" in gen_insight:
                score = gen_insight["admin_published_score"])
                return(score)
            else:
                return ("MISSING")
        else:
            return ("MISSING")


def reformat_and_clean_potentials([list_of_potentials]):
    placements_scores_convos = {}
    for record in list_of_potentials:
        app_id = record["applicant_id"]        #ID, Placement, and Score, next 3 lines
        placement_id = record["placement_id"]
        app_score = get_admin_score(record)
        
        # formatting needed to get human_only replies
        app_id_str = str(app_id)
        applicant_auth_ident = "applicant_" + app_id_str
        curr_discussion = record["discussion"]
        
        app_replies = get_clean_replies(curr_discussion, applicant_auth_ident)
        
        # make new dictionary entry of form {ID_number: {"ID":_, "placement_ID":_, 
        #                                   "applicant_score": __, "applicant_replies":__}}
        
        applicant_cleaned_info = {"ID":app_id, "placement_ID":placement_id,
                                 "applicant_score":app_score, "applicant_replies":app_replies}
        placements_scores[app_id] = applicant_cleaned_info



#get_applicant_ID
#get_placement_num
#get_score
#get_conversation
```

```python
# Function to take a document as a list of words and return the document vector
# from class code

def get_doc_vec(words, model):
    good_words = []
    for word in words:
        # Words not in the original model will fail
        try:
            if model.wv[word] is not None:
                good_words.append(word)
        except:
            continue
    # If no words are in the original model
    if len(good_words) == 0:
        return None
    # Return the mean of the vectors for all the good words
    return model.wv[good_words].mean(axis=0)
```
